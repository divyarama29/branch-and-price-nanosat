```julia
using JuMP
using Gurobi

# RMP(l) solver (with branching constraints)
function solve_RMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 1)

    @variable(model, ξ[j in J, k in K_star[j]] >= 0)
    @variable(model, 0 <= α[t in T] <= 1)
    @variable(model, b[t in T])
    @variable(model, i[t in T])
    @variable(model, ρ <= SoC[t in T] <= 1)

    @constraint(model, SoC[1] == SoC_1)

    # Objective: Include penalty for artificial columns if root node
    if is_root
        @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]) - M * sum(ξ[j, 1] for j in J))
    else
        @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]))
    end

    @constraint(model, select_profile[j in J], sum(ξ[j, k] for k in K_star[j]) == 1)
    @constraint(model, power_limit[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + γ * V_b * α[t] <= r_t[t] + γ * V_b)
    @constraint(model, power_balance[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + b[t] == r_t[t])
    @constraint(model, battery_current[t in T], i[t] == b[t] / V_b)
    @constraint(model, soc_update[t in T[1:end-1]], SoC[t+1] == SoC[t] + i[t] * e / (60 * Q))
    @constraint(model, soc_upper, SoC[T[end]] <= SoC_1 * (1 + ΔSoC))
    @constraint(model, soc_lower, SoC[T[end]] >= SoC_1 * (1 - ΔSoC))
    @constraint(model, branch_zero[(j, t) in L0], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 0)
    @constraint(model, branch_one[(j, t) in L1], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 1)

    optimize!(model)

    # println("Objective value: ", objective_value(model))
    println("Status: ", termination_status(model))

    if termination_status(model) == MOI.OPTIMAL
        ξ_val = value.(ξ)
        α_val = value.(α)
        b_val = value.(b)
        i_val = value.(i)
        SoC_val = value.(SoC)
        μ = -dual.(select_profile)
        π = -dual.(power_limit)
        ν = -dual.(power_balance)
        ζ0 = Dict((j, t) => -dual(branch_zero[(j, t)]) for (j, t) in L0)
        ζ1 = Dict((j, t) => -dual(branch_one[(j, t)]) for (j, t) in L1)
        obj_value = objective_value(model)

        return Dict(
            "status" => "Optimal",
            "objective" => obj_value,
            "ξ" => ξ_val,
            "α" => α_val,
            "b" => b_val,
            "i" => i_val,
            "SoC" => SoC_val,
            "μ" => μ,
            "π" => π,
            "ν" => ν,
            "ζ0" => ζ0,
            "ζ1" => ζ1
        )
    else
        return Dict("status" => "Infeasible")
    end
end

# Placeholder for Pricing Subproblem (P S_j)
function solve_PS_j(j, T, u_j, q_j, μ_j, π, ν, ζ0, ζ1, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0)  # Suppress solver output

    # Decision variables
    @variable(model, x[t in T], Bin)  # x_jt: Task j active at time t
    @variable(model, φ[t in T], Bin)  # φ_jt: Indicates startup of task j at time t

    # Objective function (11a)
    @objective(model, Max,
        u_j[j] * sum(x[t] for t in T) - μ_j[j] -
        q_j[j] * sum((π[t] + ν[t]) * x[t] for t in T) -
        sum(ζ0[(j, t)] * x[t] for (j, t) in keys(ζ0)) -
        sum(ζ1[(j, t)] * x[t] for (j, t) in keys(ζ1))
    )

    # Constraints (2j–2x) for task j
    # (2j): φ_j1 ≥ x_j1
    @constraint(model, φ[T[1]] >= x[T[1]])

    # (2k): φ_jt ≥ x_jt - x_j(t-1), ∀ t ∈ T \ {1}
    @constraint(model, [t in T[2:end]], φ[t] >= x[t] - x[t-1])

    # (2l): φ_jt ≤ x_jt, ∀ t ∈ T
    @constraint(model, [t in T], φ[t] <= x[t])

    # (2m): φ_jt ≤ 2 - x_jt - x_j(t-1), ∀ t ∈ T \ {1}
    @constraint(model, [t in T[2:end]], φ[t] <= 2 - x[t] - x[t-1])

    # (2n): ∑_t φ_jt ≥ y_min_j
    @constraint(model, sum(φ[t] for t in T) >= min_statup[j])

    # (2o): ∑_t φ_jt ≤ y_max_j
    @constraint(model, sum(φ[t] for t in T) <= max_statup[j])

    # (2p): ∑_{l=t}^{t+t_min_j-1} x_jl ≥ t_min_j φ_jt, ∀ t ∈ {1, ..., T-t_min_j+1}
    @constraint(model, [t in 1:(T[end] - min_cpu_time[j] + 1)],
        sum(x[l] for l in t:(t + min_cpu_time[j] - 1)) >= min_cpu_time[j] * φ[t])

    # (2q): ∑_{l=t}^{t+t_max_j} x_jl ≤ t_max_j, ∀ t ∈ {1, ..., T-t_max_j}
    @constraint(model, [t in 1:(T[end] - max_cpu_time[j])],
        sum(x[l] for l in t:(t + max_cpu_time[j])) <= max_cpu_time[j])

    # (2r): ∑_{l=t}^T x_jl ≥ (T-t+1) φ_jt, ∀ t ∈ {T-t_min_j+2, ..., T}
    @constraint(model, [t in (T[end] - min_cpu_time[j] + 2):T[end]],
        sum(x[l] for l in t:T[end]) >= (T[end] - t + 1) * φ[t])

    # (2s): ∑_{l=t}^{t+p_min_j-1} φ_jl ≤ 1, ∀ t ∈ {1 + w_min_j, ..., w_max_j - p_min_j}
    @constraint(model, [t in (1 + win_min[j]):(win_max[j] - min_periodo_job[j])],
        sum(φ[l] for l in t:(t + min_periodo_job[j] - 1)) <= 1)

    # (2t): ∑_{l=t}^{t+p_max_j-1} φ_jl ≥ 1, ∀ t ∈ {1 + w_min_j, ..., w_max_j - p_max_j}
    @constraint(model, [t in (1 + win_min[j]):(win_max[j] - max_periodo_job[j])],
        sum(φ[l] for l in t:(t + max_periodo_job[j] - 1)) >= 1)

    # (2u): x_jt = 0, t = 1, ..., w_min_j
    @constraint(model, [t in 1:win_min[j]], x[t] == 0)

    # (2v): x_jt = 0, t = w_max_j, ..., T
    @constraint(model, [t in win_max[j]:T[end]], x[t] == 0)

    # (2w–2x): x_jt, φ_jt binary (already enforced by variable definitions)

    # Solve the model
    optimize!(model)

    if termination_status(model) == MOI.OPTIMAL
        x_val = value.(x)
        reduced_cost = objective_value(model)
        profile = [round(Int, x_val[t]) for t in T]
        if reduced_cost > 1e-6  # Small tolerance for numerical stability
            return Dict("reduced_cost" => reduced_cost, "profile" => profile)
        else
            return Dict("reduced_cost" => reduced_cost)
        end
    else
        return Dict("reduced_cost" => -Inf)
    end
end

# Placeholder for MIP Heuristic (HRMP)
function solve_HRMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1)
    # Initialize the model
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0)  # Suppress solver output
    set_optimizer_attribute(model, "TimeLimit", 60)  # Set time limit (e.g., 60 seconds)
    set_optimizer_attribute(model, "MIPGap", 0.05)   # Set MIP gap tolerance to 5%

    # Decision variables
    @variable(model, ξ[j in J, k in K_star[j]], Bin)  # Binary profile selection variables
    @variable(model, 0 <= α[t in T] <= 1)            # Battery energy drain
    @variable(model, b[t in T])                      # Power balance
    @variable(model, i[t in T])                      # Battery current
    @variable(model, ρ <= SoC[t in T] <= 1)          # State of charge

    # Set initial SoC
    @constraint(model, SoC[1] == SoC_1)

    # Objective: Maximize mission value
    @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]))

    # Constraints
    # (5b): Exactly one profile per task
    @constraint(model, select_profile[j in J],
        sum(ξ[j, k] for k in K_star[j]) == 1)

    # (5c): Power consumption limit
    @constraint(model, power_limit[t in T],
        sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + γ * V_b * α[t] <= r_t[t] + γ * V_b)

    # (5d): Power balance
    @constraint(model, power_balance[t in T],
        sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + b[t] == r_t[t])

    # (2d): Battery current
    @constraint(model, battery_current[t in T],
        i[t] == b[t] / V_b)

    # (2e): State of charge update
    @constraint(model, soc_update[t in T[1:end-1]],
        SoC[t+1] == SoC[t] + i[t] * e / (60 * Q))

    # (2h) and (2i): SoC at end of orbit close to start
    @constraint(model, soc_upper,
        SoC[T[end]] <= SoC_1 * (1 + ΔSoC))
    @constraint(model, soc_lower,
        SoC[T[end]] >= SoC_1 * (1 - ΔSoC))

    # (10b): Branching constraints for x_jt = 0
    @constraint(model, branch_zero[(j, t) in L0],
        sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 0)

    # (10c): Branching constraints for x_jt = 1
    @constraint(model, branch_one[(j, t) in L1],
        sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 1)

    # Solve the model
    optimize!(model)

    # Check solution status
    if termination_status(model) == MOI.OPTIMAL || termination_status(model) == MOI.TIME_LIMIT
        if has_values(model)
            ξ_val = value.(ξ)
            obj_value = objective_value(model)
            return Dict(
                "feasible" => true,
                "objective" => obj_value,
                "ξ" => ξ_val
            )
        else
            return Dict("feasible" => false)
        end
    else
        return Dict("feasible" => false)
    end
end

# Node processing function (Algorithm 1)
function node(l, LB, UB_l, J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    # Column generation loop
    new_columns_added = true
    while new_columns_added
        # Line 3: Solve RMP(l)
        println("K_star",K_star)
        println("θ_k",θ_k)
        rmp_result = solve_RMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root)
        println("RMP status: ", rmp_result["status"])

        # Line 4–5: Check infeasibility
        if rmp_result["status"] == "Infeasible"
            return 0, LB, UB_l, [], K_star, θ_k, c_k, q_k
        end

        Z_RMP = rmp_result["objective"]
        σ = Dict("μ" => rmp_result["μ"], "π" => rmp_result["π"], "ν" => rmp_result["ν"], "ζ0" => rmp_result["ζ0"], "ζ1" => rmp_result["ζ1"])

        # Line 6–9: Solve pricing subproblems
        new_columns_added = false
        max_reduced_cost = -Inf
        for j in J
            ps_result = solve_PS_j(j, T, u_j, q_j, σ["μ"][j], σ["π"], σ["ν"], σ["ζ0"], σ["ζ1"], min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
            if ps_result["reduced_cost"] > 0
                new_columns_added = true
                max_reduced_cost = max(max_reduced_cost, ps_result["reduced_cost"])
                # Add new profile to K_star, θ_k, c_k, q_k
                k_new = length(K_star[j]) + 1
                K_star[j] = push!(K_star[j], k_new)
                θ_k[j][k_new] = ps_result["profile"]
                c_k[j][k_new] = u_j[j] * sum(ps_result["profile"])
                q_k[j][k_new] = [q_j[j] * ps_result["profile"][t] for t in T]
            end
        end
    end

    println( "Theta k",θ_k)

    # Line 11–12: Prune if Z_RMP ≤ LB
    if Z_RMP <= LB
        return 0, LB, UB_l, [], K_star, θ_k, c_k, q_k
    end

    # Line 13–14: Update upper bound
    if Z_RMP < UB_l
        UB_l = Z_RMP
    end

    # Line 15–21: Solve MIP heuristic
    hrmp_result = solve_HRMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1)
    if hrmp_result["feasible"]
        Z_H = hrmp_result["objective"]
        if Z_H > LB
            LB = Z_H
        end
        if Z_H == UB_l
            return 1, LB, UB_l, [], K_star, θ_k, c_k, q_k
        end
    end

    # Line 22–23: Find most fractional x_jt
    x_jt = Dict((j, t) => sum(θ_k[j][k][t] * rmp_result["ξ"][j, k] for k in K_star[j]) for j in J, t in T)
    min_fractional = Inf
    selected_jt = (0, 0)
    for j in J, t in T
        fractional = abs(0.5 - x_jt[j, t])
        if fractional < min_fractional
            min_fractional = fractional
            selected_jt = (j, t)
        end
    end

    # Line 24–26: Branch if fractional
    new_nodes = []
    if min_fractional < 0.499  # Consider x_jt fractional if not close to 0 or 1
        push!(new_nodes, (l * 2, copy(L0) ∪ [selected_jt], copy(L1)))  # x_jt = 0
        push!(new_nodes, (l * 2 + 1, copy(L0), copy(L1) ∪ [selected_jt]))  # x_jt = 1
    else
        # Line 27–29: Update lower bound if integer
        if Z_RMP > LB
            LB = Z_RMP
        end
    end

    return nothing, LB, UB_l, new_nodes, K_star, θ_k, c_k, q_k
    return nothing
end



# Prepare the data
# Prepare the data
J = 1:13
T = 1:120
r_t = Dict(t => recurso_p[t] for t in T)
u_j = Dict(j => priority[j] for j in J)
q_j = Dict(j => uso_p[j] for j in J)
min_statup = Dict(j => min_statup[j] for j in J)
max_statup = Dict(j => max_statup[j] for j in J)
min_cpu_time = Dict(j => min_cpu_time[j] for j in J)
max_cpu_time = Dict(j => max_cpu_time[j] for j in J)
min_periodo_job = Dict(j => min_periodo_job[j] for j in J)
max_periodo_job = Dict(j => max_periodo_job[j] for j in J)
win_min = Dict(j => win_min[j] for j in J)
win_max = Dict(j => win_max[j] for j in J)

# Battery parameters
V_b = 3.6
γ = 2.0
Q = 6.0
e = 0.9
ρ = 0.5
ΔSoC = 0.05
SoC_1 = 0.75
M = 1e6

# Initial profiles: artificial all-ones columns
K_star = Dict(j => [1] for j in J)
θ_k = Dict(j => Dict(1 => colunas_seed[j][1]) for j in J)
c_k = Dict(j => Dict{Int, Float64}() for j in J)
q_k = Dict(j => Dict{Int, Vector{Float64}}() for j in J)
for j in J
    c_k[j][1] = u_j[j] * sum(θ_k[j][1])
    q_k[j][1] = [q_j[j] * θ_k[j][1][t] for t in T]
end


# Initialize node
l = 1  # Root node
LB = -Inf
UB_l = Inf
L0 = Set{Tuple{Int,Int}}()
L1 = Set{Tuple{Int,Int}}()
is_root = true

# Process root node
result, LB, UB_l, new_nodes, K_star, θ_k, c_k, q_k = node(l, LB, UB_l, J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)

# Print results
if result == 0
    println("Node $l pruned")
elseif result == 1
    println("Optimal integer solution found at node $l")
else
    println("Node $l processed, new nodes: ", new_nodes)
    println("Updated LB: ", LB)
    println("Updated UB_l: ", UB_l)
end



# rmp_result = solve_RMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root)

# # Print results
# println("RMP(1) Status: ", rmp_result["status"])
# if rmp_result["status"] == "Optimal"
#     println("Objective Value: ", rmp_result["objective"])
#     println("Dual μ: ", rmp_result["μ"])
#     println("Dual π (first 5 periods): ", [rmp_result["π"][t] for t in 1:min(5, length(T))])
#     println("Dual ν (first 5 periods): ", [rmp_result["ν"][t] for t in 1:min(5, length(T))])
# else
#     println("RMP(1) is infeasible. Check power constraints.")
#     for t in T
#         power = sum(q_j[j] for j in J)  # Power from all-ones profiles
#         limit = r_t[t] + γ * V_b
#         if power > limit
#             println("Time $t: Power demand $power exceeds limit $limit")
#         end
#     end
# end




# # Initial profiles: artificial (all-ones) and feasible
# K_star = Dict(j => [1, 2] for j in J)  # Two profiles: 1 (artificial), 2 (feasible)
# θ_k = Dict(j => Dict{Int, Vector{Int}}() for j in J)
# c_k = Dict(j => Dict{Int, Float64}() for j in J)
# q_k = Dict(j => Dict{Int, Vector{Float64}}() for j in J)

# for j in J
#     # Artificial column (k = 1): all-ones, as per section 5.4
#     θ_k[j][1] = ones(Int, 120)
#     c_k[j][1] = u_j[j] * sum(θ_k[j][1])
#     q_k[j][1] = [q_j[j] * θ_k[j][1][t] for t in T]

#     # Feasible column (k = 2): heuristic to satisfy constraints (2j–2x)
#     profile = zeros(Int, 120)
#     active_periods = min_cpu_time[j]
#     startups_needed = min(min_statup[j], max_statup[j])  # Ensure within bounds
#     period_gap = min_periodo_job[j]
    
#     # Place startups and active periods
#     t = win_min[j] + 1  # Start within window
#     startups = 0
#     periods_assigned = 0
    
#     while startups < startups_needed && periods_assigned < active_periods && t <= win_max[j]
#         # Place a startup
#         profile[t] = 1
#         periods_assigned += 1
#         startups += 1
        
#         # Add additional periods to meet min_cpu_time if needed
#         if periods_assigned < active_periods && t < win_max[j]
#             profile[t + 1] = 1
#             periods_assigned += 1
#             t += 1
#         end
        
#         # Move to next possible startup time
#         t += period_gap
#     end
    
#     # Fill remaining periods to meet min_cpu_time if needed
#     while periods_assigned < active_periods && t <= win_max[j]
#         profile[t] = 1
#         periods_assigned += 1
#         t += 1
#     end
    
#     # Store the feasible profile
#     θ_k[j][2] = profile
#     c_k[j][2] = u_j[j] * sum(profile)
#     q_k[j][2] = [q_j[j] * profile[t] for t in T]
# end
```

```julia
using JuMP, Gurobi, DataStructures

# Fallback data (replace with actual data)
J = 1:13
T = 1:120
recurso_p = Dict(t => 10.0 for t in T)  # Example: constant resource
priority = Dict(j => 10 + j for j in J)  # Example: increasing priority
uso_p = Dict(j => 1.0 for j in J)  # Example: constant usage
min_statup = Dict(j => 9 for j in J)
max_statup = Dict(j => 15 for j in J)
min_cpu_time = Dict(j => 2 for j in J)
max_cpu_time = Dict(j => 5 for j in J)
min_periodo_job = Dict(j => 5 for j in J)
max_periodo_job = Dict(j => 10 for j in J)
win_min = Dict(j => 0 for j in J)
win_max = Dict(j => 120 for j in J)
colunas_seed = Dict(j => [ones(Int, 120)] for j in J)  # Fallback: all-ones profile
V_b = 3.6
γ = 2.0
Q = 6.0
e = 0.9
ρ = 0.5
ΔSoC = 0.05
SoC_1 = 0.75
M = 1e6

# Derived data
r_t = Dict(t => recurso_p[t] for t in T)
u_j = Dict(j => priority[j] for j in J)
q_j = Dict(j => uso_p[j] for j in J)
min_statup = Dict(j => min_statup[j] for j in J)
max_statup = Dict(j => max_statup[j] for j in J)
min_cpu_time = Dict(j => min_cpu_time[j] for j in J)
max_cpu_time = Dict(j => max_cpu_time[j] for j in J)
min_periodo_job = Dict(j => min_periodo_job[j] for j in J)
max_periodo_job = Dict(j => max_periodo_job[j] for j in J)
win_min = Dict(j => win_min[j] for j in J)
win_max = Dict(j => win_max[j] for j in J)

# Initial profiles
K_star = Dict(j => [1] for j in J)
θ_k = Dict(j => Dict(1 => colunas_seed[j][1]) for j in J)
c_k = Dict(j => Dict{Int, Float64}() for j in J)
q_k = Dict(j => Dict{Int, Vector{Float64}}() for j in J)
for j in J
    c_k[j][1] = u_j[j] * sum(θ_k[j][1])
    q_k[j][1] = [q_j[j] * θ_k[j][1][t] for t in T]
end

# Node structure
mutable struct Node
    id::Int
    UB::Float64
    L0::Set{Tuple{Int,Int}}
    L1::Set{Tuple{Int,Int}}
    is_root::Bool
end

# solve_RMP_l
function solve_RMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 1)

    @variable(model, ξ[j in J, k in K_star[j]] >= 0)
    @variable(model, 0 <= α[t in T] <= 1)
    @variable(model, b[t in T])
    @variable(model, i[t in T])
    @variable(model, ρ <= SoC[t in T] <= 1)

    @constraint(model, SoC[1] == SoC_1)

    if is_root
        @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]) - M * sum(ξ[j, 1] for j in J))
    else
        @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]))
    end

    @constraint(model, select_profile[j in J], sum(ξ[j, k] for k in K_star[j]) == 1)
    @constraint(model, power_limit[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + γ * V_b * α[t] <= r_t[t] + γ * V_b)
    @constraint(model, power_balance[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + b[t] == r_t[t])
    @constraint(model, battery_current[t in T], i[t] == b[t] / V_b)
    @constraint(model, soc_update[t in T[1:end-1]], SoC[t+1] == SoC[t] + i[t] * e / (60 * Q))
    @constraint(model, soc_upper, SoC[T[end]] <= SoC_1 * (1 + ΔSoC))
    @constraint(model, soc_lower, SoC[T[end]] >= SoC_1 * (1 - ΔSoC))
    @constraint(model, branch_zero[(j, t) in L0], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 0)
    @constraint(model, branch_one[(j, t) in L1], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 1)

    optimize!(model)
    println("Status: ", termination_status(model))

    if termination_status(model) == MOI.OPTIMAL
        ξ_val = value.(ξ)
        α_val = value.(α)
        b_val = value.(b)
        i_val = value.(i)
        SoC_val = value.(SoC)
        μ = -dual.(select_profile)
        π = -dual.(power_limit)
        ν = -dual.(power_balance)
        ζ0 = Dict((j, t) => -dual(branch_zero[(j, t)]) for (j, t) in L0)
        ζ1 = Dict((j, t) => -dual(branch_one[(j, t)]) for (j, t) in L1)
        obj_value = objective_value(model)

        return Dict(
            "status" => "Optimal",
            "objective" => obj_value,
            "ξ" => ξ_val,
            "α" => α_val,
            "b" => b_val,
            "i" => i_val,
            "SoC" => SoC_val,
            "μ" => μ,
            "π" => π,
            "ν" => ν,
            "ζ0" => ζ0,
            "ζ1" => ζ1
        )
    else
        return Dict("status" => "Infeasible")
    end
end

# solve_PS_j
function solve_PS_j(j, T, u_j, q_j, μ_j, π, ν, ζ0, ζ1, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0)

    @variable(model, x[t in T], Bin)
    @variable(model, φ[t in T], Bin)

    @objective(model, Max,
        u_j[j] * sum(x[t] for t in T) - μ_j[j] -
        q_j[j] * sum((π[t] + ν[t]) * x[t] for t in T) -
        sum(ζ0[(j, t)] * x[t] for (j, t) in keys(ζ0)) -
        sum(ζ1[(j, t)] * x[t] for (j, t) in keys(ζ1))
    )

    @constraint(model, φ[T[1]] >= x[T[1]])
    @constraint(model, [t in T[2:end]], φ[t] >= x[t] - x[t-1])
    @constraint(model, [t in T], φ[t] <= x[t])
    @constraint(model, [t in T[2:end]], φ[t] <= 2 - x[t] - x[t-1])
    @constraint(model, sum(φ[t] for t in T) >= min_statup[j])
    @constraint(model, sum(φ[t] for t in T) <= max_statup[j])
    @constraint(model, [t in 1:(T[end] - min_cpu_time[j] + 1)],
        sum(x[l] for l in t:(t + min_cpu_time[j] - 1)) >= min_cpu_time[j] * φ[t])
    @constraint(model, [t in 1:(T[end] - max_cpu_time[j])],
        sum(x[l] for l in t:(t + max_cpu_time[j])) <= max_cpu_time[j])
    @constraint(model, [t in (T[end] - min_cpu_time[j] + 2):T[end]],
        sum(x[l] for l in t:T[end]) >= (T[end] - t + 1) * φ[t])
    @constraint(model, [t in (1 + win_min[j]):(win_max[j] - min_periodo_job[j])],
        sum(φ[l] for l in t:(t + min_periodo_job[j] - 1)) <= 1)
    @constraint(model, [t in (1 + win_min[j]):(win_max[j] - max_periodo_job[j])],
        sum(φ[l] for l in t:(t + max_periodo_job[j] - 1)) >= 1)
    @constraint(model, [t in 1:win_min[j]], x[t] == 0)
    @constraint(model, [t in win_max[j]:T[end]], x[t] == 0)

    optimize!(model)

    if termination_status(model) == MOI.OPTIMAL
        x_val = value.(x)
        reduced_cost = objective_value(model)
        profile = [round(Int, x_val[t]) for t in T]
        if reduced_cost > 1e-6
            return Dict("reduced_cost" => reduced_cost, "profile" => profile)
        else
            return Dict("reduced_cost" => reduced_cost)
        end
    else
        return Dict("reduced_cost" => -Inf)
    end
end

# solve_HRMP_l
function solve_HRMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0)
    set_optimizer_attribute(model, "TimeLimit", 60)
    set_optimizer_attribute(model, "MIPGap", 0.05)

    @variable(model, ξ[j in J, k in K_star[j]], Bin)
    @variable(model, 0 <= α[t in T] <= 1)
    @variable(model, b[t in T])
    @variable(model, i[t in T])
    @variable(model, ρ <= SoC[t in T] <= 1)

    @constraint(model, SoC[1] == SoC_1)
    @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]))

    @constraint(model, select_profile[j in J], sum(ξ[j, k] for k in K_star[j]) == 1)
    @constraint(model, power_limit[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + γ * V_b * α[t] <= r_t[t] + γ * V_b)
    @constraint(model, power_balance[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + b[t] == r_t[t])
    @constraint(model, battery_current[t in T], i[t] == b[t] / V_b)
    @constraint(model, soc_update[t in T[1:end-1]], SoC[t+1] == SoC[t] + i[t] * e / (60 * Q))
    @constraint(model, soc_upper, SoC[T[end]] <= SoC_1 * (1 + ΔSoC))
    @constraint(model, soc_lower, SoC[T[end]] >= SoC_1 * (1 - ΔSoC))
    @constraint(model, branch_zero[(j, t) in L0], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 0)
    @constraint(model, branch_one[(j, t) in L1], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 1)

    optimize!(model)

    if termination_status(model) == MOI.OPTIMAL || termination_status(model) == MOI.TIME_LIMIT
        if has_values(model)
            ξ_val = value.(ξ)
            obj_value = objective_value(model)
            return Dict("feasible" => true, "objective" => obj_value, "ξ" => ξ_val)
        else
            return Dict("feasible" => false)
        end
    else
        return Dict("feasible" => false)
    end
end

# node
function node(l, LB, UB_l, J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    new_columns_added = true
    iteration = 0
    while new_columns_added
        iteration += 1
        println("Column generation iteration $iteration for node $l")
        rmp_result = solve_RMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root)
        println("RMP status: ", rmp_result["status"])

        if rmp_result["status"] == "Infeasible"
            return 0, LB, UB_l, [], K_star, θ_k, c_k, q_k
        end

        Z_RMP = rmp_result["objective"]
        σ = Dict("μ" => rmp_result["μ"], "π" => rmp_result["π"], "ν" => rmp_result["ν"], "ζ0" => rmp_result["ζ0"], "ζ1" => rmp_result["ζ1"])

        new_columns_added = false
        max_reduced_cost = -Inf
        for j in J
            ps_result = solve_PS_j(j, T, u_j, q_j, σ["μ"][j], σ["π"], σ["ν"], σ["ζ0"], σ["ζ1"], min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
            if ps_result["reduced_cost"] > 0
                new_columns_added = true
                max_reduced_cost = max(max_reduced_cost, ps_result["reduced_cost"])
                k_new = length(K_star[j]) + 1
                K_star[j] = push!(K_star[j], k_new)
                θ_k[j][k_new] = ps_result["profile"]
                c_k[j][k_new] = u_j[j] * sum(ps_result["profile"])
                q_k[j][k_new] = [q_j[j] * ps_result["profile"][t] for t in T]
                println("Task $j: Added profile $k_new, Reduced cost = $(ps_result["reduced_cost"]), Active periods = $(sum(ps_result["profile"]))")
            end
        end
    end

    if Z_RMP <= LB
        return 0, LB, UB_l, [], K_star, θ_k, c_k, q_k
    end

    if Z_RMP < UB_l
        UB_l = Z_RMP
    end

    hrmp_result = solve_HRMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1)
    if hrmp_result["feasible"]
        Z_H = hrmp_result["objective"]
        if Z_H > LB
            LB = Z_H
        end
        if Z_H == UB_l
            return 1, LB, UB_l, [], K_star, θ_k, c_k, q_k
        end
    end

    x_jt = Dict((j, t) => sum(θ_k[j][k][t] * rmp_result["ξ"][j, k] for k in K_star[j]) for j in J, t in T)
    min_fractional = Inf
    selected_jt = (0, 0)
    for j in J, t in T
        fractional = abs(0.5 - x_jt[j, t])
        if fractional < min_fractional
            min_fractional = fractional
            selected_jt = (j, t)
        end
    end

    new_nodes = []
    if min_fractional < 0.499
        push!(new_nodes, (l * 2, copy(L0) ∪ [selected_jt], copy(L1)))  # x_jt = 0
        push!(new_nodes, (l * 2 + 1, copy(L0), copy(L1) ∪ [selected_jt]))  # x_jt = 1
        return 2, LB, UB_l, new_nodes, K_star, θ_k, c_k, q_k
    else
        if Z_RMP > LB
            LB = Z_RMP
        end
        return 1, LB, UB_l, [], K_star, θ_k, c_k, q_k
    end
end

# branch_and_price
function branch_and_price(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, M, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    nodes = PriorityQueue{Int, Float64}(Base.Order.Reverse)
    node_data = Dict{Int, Node}()
    l = 1
    LB = -Inf
    UB_l = Inf
    L0 = Set{Tuple{Int,Int}}()
    L1 = Set{Tuple{Int,Int}}()
    is_root = true
    node_data[l] = Node(l, UB_l, L0, L1, is_root)
    enqueue!(nodes, l => UB_l)

    while !isempty(nodes)
        l, UB_l = dequeue_pair!(nodes)
        current_node = node_data[l]  # Renamed to avoid conflict
        println("Processing node $(current_node.id) with UB = $(current_node.UB)")

        result, LB, UB_l, new_nodes, K_star, θ_k, c_k, q_k = node(  # Call the node function
            current_node.id, LB, current_node.UB, J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, current_node.L0, current_node.L1, M, current_node.is_root,
            min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max
        )

        if result == 0
            println("Node $(current_node.id) pruned")
        elseif result == 1
            println("Optimal integer solution found at node $(current_node.id)")
            return K_star, θ_k, c_k, q_k, LB
        else
            println("Node $(current_node.id) processed, new nodes: ", [n[1] for n in new_nodes])
            println("Updated LB: $LB, UB: $UB_l")
            for (new_l, new_L0, new_L1) in new_nodes
                new_UB = UB_l
                if new_UB > LB + 1e-6
                    node_data[new_l] = Node(new_l, new_UB, new_L0, new_L1, false)
                    enqueue!(nodes, new_l => new_UB)
                end
            end
        end
    end

    println("No feasible solution found")
    return K_star, θ_k, c_k, q_k, LB
end

# Run
K_star, θ_k, c_k, q_k, LB = branch_and_price(
    J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, M,
    min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max
)

# Print final results
println("Final LB: ", LB)
println("Number of profiles per task: ", Dict(j => length(K_star[j]) for j in J))
```

```julia
using JuMP, Gurobi, DataStructures

# Fallback data (replace with actual data)
J = 1:13
T = 1:120
recurso_p = Dict(t => 10.0 for t in T)  # Example: constant resource
priority = Dict(j => 10 + j for j in J)  # Example: increasing priority
uso_p = Dict(j => 1.0 for j in J)  # Example: constant usage
min_statup = Dict(j => 9 for j in J)
max_statup = Dict(j => 15 for j in J)
min_cpu_time = Dict(j => 2 for j in J)
max_cpu_time = Dict(j => 5 for j in J)
min_periodo_job = Dict(j => 5 for j in J)
max_periodo_job = Dict(j => 10 for j in J)
win_min = Dict(j => 0 for j in J)
win_max = Dict(j => 120 for j in J)
colunas_seed = Dict(j => [ones(Int, 120)] for j in J)  # Fallback: all-ones profile
V_b = 3.6
γ = 2.0
Q = 6.0
e = 0.9
ρ = 0.5
ΔSoC = 0.05
SoC_1 = 0.75
M = 1e6

# Derived data
r_t = Dict(t => recurso_p[t] for t in T)
u_j = Dict(j => priority[j] for j in J)
q_j = Dict(j => uso_p[j] for j in J)
min_statup = Dict(j => min_statup[j] for j in J)
max_statup = Dict(j => max_statup[j] for j in J)
min_cpu_time = Dict(j => min_cpu_time[j] for j in J)
max_cpu_time = Dict(j => max_cpu_time[j] for j in J)
min_periodo_job = Dict(j => min_periodo_job[j] for j in J)
max_periodo_job = Dict(j => max_periodo_job[j] for j in J)
win_min = Dict(j => win_min[j] for j in J)
win_max = Dict(j => win_max[j] for j in J)

# Initial profiles
K_star = Dict(j => [1] for j in J)
θ_k = Dict(j => Dict(1 => colunas_seed[j][1]) for j in J)
c_k = Dict(j => Dict{Int, Float64}() for j in J)
q_k = Dict(j => Dict{Int, Vector{Float64}}() for j in J)
for j in J
    c_k[j][1] = u_j[j] * sum(θ_k[j][1])
    q_k[j][1] = [q_j[j] * θ_k[j][1][t] for t in T]
end

# Node structure
mutable struct Node
    id::Int
    UB::Float64
    L0::Set{Tuple{Int,Int}}
    L1::Set{Tuple{Int,Int}}
    is_root::Bool
end

# solve_RMP_l
function solve_RMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 1)

    @variable(model, ξ[j in J, k in K_star[j]] >= 0)
    @variable(model, 0 <= α[t in T] <= 1)
    @variable(model, b[t in T])
    @variable(model, i[t in T])
    @variable(model, ρ <= SoC[t in T] <= 1)

    @constraint(model, SoC[1] == SoC_1)

    if is_root
        @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]) - M * sum(ξ[j, 1] for j in J))
    else
        @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]))
    end

    @constraint(model, select_profile[j in J], sum(ξ[j, k] for k in K_star[j]) == 1)
    @constraint(model, power_limit[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + γ * V_b * α[t] <= r_t[t] + γ * V_b)
    @constraint(model, power_balance[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + b[t] == r_t[t])
    @constraint(model, battery_current[t in T], i[t] == b[t] / V_b)
    @constraint(model, soc_update[t in T[1:end-1]], SoC[t+1] == SoC[t] + i[t] * e / (60 * Q))
    @constraint(model, soc_upper, SoC[T[end]] <= SoC_1 * (1 + ΔSoC))
    @constraint(model, soc_lower, SoC[T[end]] >= SoC_1 * (1 - ΔSoC))
    @constraint(model, branch_zero[(j, t) in L0], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 0)
    @constraint(model, branch_one[(j, t) in L1], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 1)

    optimize!(model)
    println("Status: ", termination_status(model))

    if termination_status(model) == MOI.OPTIMAL
        ξ_val = value.(ξ)
        α_val = value.(α)
        b_val = value.(b)
        i_val = value.(i)
        SoC_val = value.(SoC)
        μ = -dual.(select_profile)
        π = -dual.(power_limit)
        ν = -dual.(power_balance)
        ζ0 = Dict((j, t) => -dual(branch_zero[(j, t)]) for (j, t) in L0)
        ζ1 = Dict((j, t) => -dual(branch_one[(j, t)]) for (j, t) in L1)
        obj_value = objective_value(model)

        return Dict(
            "status" => "Optimal",
            "objective" => obj_value,
            "ξ" => ξ_val,
            "α" => α_val,
            "b" => b_val,
            "i" => i_val,
            "SoC" => SoC_val,
            "μ" => μ,
            "π" => π,
            "ν" => ν,
            "ζ0" => ζ0,
            "ζ1" => ζ1
        )
    else
        return Dict("status" => "Infeasible")
    end
end

# solve_PS_j
function solve_PS_j(j, T, u_j, q_j, μ_j, π, ν, ζ0, ζ1, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0)

    @variable(model, x[t in T], Bin)
    @variable(model, φ[t in T], Bin)

    @objective(model, Max,
        u_j[j] * sum(x[t] for t in T) - μ_j[j] -
        q_j[j] * sum((π[t] + ν[t]) * x[t] for t in T) -
        sum(ζ0[(j, t)] * x[t] for (j, t) in keys(ζ0)) -
        sum(ζ1[(j, t)] * x[t] for (j, t) in keys(ζ1))
    )

    @constraint(model, φ[T[1]] >= x[T[1]])
    @constraint(model, [t in T[2:end]], φ[t] >= x[t] - x[t-1])
    @constraint(model, [t in T], φ[t] <= x[t])
    @constraint(model, [t in T[2:end]], φ[t] <= 2 - x[t] - x[t-1])
    @constraint(model, sum(φ[t] for t in T) >= min_statup[j])
    @constraint(model, sum(φ[t] for t in T) <= max_statup[j])
    @constraint(model, [t in 1:(T[end] - min_cpu_time[j] + 1)],
        sum(x[l] for l in t:(t + min_cpu_time[j] - 1)) >= min_cpu_time[j] * φ[t])
    @constraint(model, [t in 1:(T[end] - max_cpu_time[j])],
        sum(x[l] for l in t:(t + max_cpu_time[j])) <= max_cpu_time[j])
    @constraint(model, [t in (T[end] - min_cpu_time[j] + 2):T[end]],
        sum(x[l] for l in t:T[end]) >= (T[end] - t + 1) * φ[t])
    @constraint(model, [t in (1 + win_min[j]):(win_max[j] - min_periodo_job[j])],
        sum(φ[l] for l in t:(t + min_periodo_job[j] - 1)) <= 1)
    @constraint(model, [t in (1 + win_min[j]):(win_max[j] - max_periodo_job[j])],
        sum(φ[l] for l in t:(t + max_periodo_job[j] - 1)) >= 1)
    @constraint(model, [t in 1:win_min[j]], x[t] == 0)
    @constraint(model, [t in win_max[j]:T[end]], x[t] == 0)

    optimize!(model)

    if termination_status(model) == MOI.OPTIMAL
        x_val = value.(x)
        reduced_cost = objective_value(model)
        profile = [round(Int, x_val[t]) for t in T]
        if reduced_cost > 1e-6
            return Dict("reduced_cost" => reduced_cost, "profile" => profile)
        else
            return Dict("reduced_cost" => reduced_cost)
        end
    else
        return Dict("reduced_cost" => -Inf)
    end
end

# solve_HRMP_l
function solve_HRMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1)
    model = Model(Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0)
    set_optimizer_attribute(model, "TimeLimit", 60)
    set_optimizer_attribute(model, "MIPGap", 0.05)

    @variable(model, ξ[j in J, k in K_star[j]], Bin)
    @variable(model, 0 <= α[t in T] <= 1)
    @variable(model, b[t in T])
    @variable(model, i[t in T])
    @variable(model, ρ <= SoC[t in T] <= 1)

    @constraint(model, SoC[1] == SoC_1)
    @objective(model, Max, sum(c_k[j][k] * ξ[j, k] for j in J, k in K_star[j]))

    @constraint(model, select_profile[j in J], sum(ξ[j, k] for k in K_star[j]) == 1)
    @constraint(model, power_limit[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + γ * V_b * α[t] <= r_t[t] + γ * V_b)
    @constraint(model, power_balance[t in T], sum(q_k[j][k][t] * ξ[j, k] for j in J, k in K_star[j]) + b[t] == r_t[t])
    @constraint(model, battery_current[t in T], i[t] == b[t] / V_b)
    @constraint(model, soc_update[t in T[1:end-1]], SoC[t+1] == SoC[t] + i[t] * e / (60 * Q))
    @constraint(model, soc_upper, SoC[T[end]] <= SoC_1 * (1 + ΔSoC))
    @constraint(model, soc_lower, SoC[T[end]] >= SoC_1 * (1 - ΔSoC))
    @constraint(model, branch_zero[(j, t) in L0], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 0)
    @constraint(model, branch_one[(j, t) in L1], sum(θ_k[j][k][t] * ξ[j, k] for k in K_star[j]) == 1)

    optimize!(model)

    if termination_status(model) == MOI.OPTIMAL || termination_status(model) == MOI.TIME_LIMIT
        if has_values(model)
            ξ_val = value.(ξ)
            obj_value = objective_value(model)
            return Dict("feasible" => true, "objective" => obj_value, "ξ" => ξ_val)
        else
            return Dict("feasible" => false)
        end
    else
        return Dict("feasible" => false)
    end
end

# node
function node(l, LB, UB_l, J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    new_columns_added = true
    iteration = 0
    while new_columns_added
        iteration += 1
        println("Column generation iteration $iteration for node $l")
        rmp_result = solve_RMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1, M, is_root)
        println("RMP status: ", rmp_result["status"])

        if rmp_result["status"] == "Infeasible"
            return 0, LB, UB_l, [], K_star, θ_k, c_k, q_k
        end

        Z_RMP = rmp_result["objective"]
        σ = Dict("μ" => rmp_result["μ"], "π" => rmp_result["π"], "ν" => rmp_result["ν"], "ζ0" => rmp_result["ζ0"], "ζ1" => rmp_result["ζ1"])

        new_columns_added = false
        max_reduced_cost = -Inf
        for j in J
            ps_result = solve_PS_j(j, T, u_j, q_j, σ["μ"][j], σ["π"], σ["ν"], σ["ζ0"], σ["ζ1"], min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
            if ps_result["reduced_cost"] > 0
                new_columns_added = true
                max_reduced_cost = max(max_reduced_cost, ps_result["reduced_cost"])
                k_new = length(K_star[j]) + 1
                K_star[j] = push!(K_star[j], k_new)
                θ_k[j][k_new] = ps_result["profile"]
                c_k[j][k_new] = u_j[j] * sum(ps_result["profile"])
                q_k[j][k_new] = [q_j[j] * ps_result["profile"][t] for t in T]
                println("Task $j: Added profile $k_new, Reduced cost = $(ps_result["reduced_cost"]), Active periods = $(sum(ps_result["profile"]))")
            end
        end
    end

    if Z_RMP <= LB
        return 0, LB, UB_l, [], K_star, θ_k, c_k, q_k
    end

    if Z_RMP < UB_l
        UB_l = Z_RMP
    end

    hrmp_result = solve_HRMP_l(J, T, K_star, θ_k, c_k, q_k, r_t, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, L0, L1)
    if hrmp_result["feasible"]
        Z_H = hrmp_result["objective"]
        if Z_H > LB
            LB = Z_H
        end
        if Z_H == UB_l
            return 1, LB, UB_l, [], K_star, θ_k, c_k, q_k
        end
    end

    x_jt = Dict((j, t) => sum(θ_k[j][k][t] * rmp_result["ξ"][j, k] for k in K_star[j]) for j in J, t in T)
    min_fractional = Inf
    selected_jt = (0, 0)
    for j in J, t in T
        fractional = abs(0.5 - x_jt[j, t])
        if fractional < min_fractional
            min_fractional = fractional
            selected_jt = (j, t)
        end
    end

    new_nodes = []
    if min_fractional < 0.499
        push!(new_nodes, (l * 2, copy(L0) ∪ [selected_jt], copy(L1)))  # x_jt = 0
        push!(new_nodes, (l * 2 + 1, copy(L0), copy(L1) ∪ [selected_jt]))  # x_jt = 1
        return 2, LB, UB_l, new_nodes, K_star, θ_k, c_k, q_k
    else
        if Z_RMP > LB
            LB = Z_RMP
        end
        return 1, LB, UB_l, [], K_star, θ_k, c_k, q_k
    end
end

# branch_and_price

function branch_and_price(J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, M, min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max)
    nodes = Vector{Int}()  # Stack for DFS (LIFO)
    node_data = Dict{Int, Node}()
    l = 1
    LB = -Inf
    UB_l = 0.0  # Placeholder, not used for prioritization
    L0 = Set{Tuple{Int,Int}}()
    L1 = Set{Tuple{Int,Int}}()
    is_root = true
    node_data[l] = Node(l, UB_l, L0, L1, is_root)
    push!(nodes, l)

    while !isempty(nodes)
        l = pop!(nodes)  # LIFO: Get most recent node
        current_node = node_data[l]
        println("Processing node $(current_node.id)")

        result, LB, UB_l, new_nodes, K_star, θ_k, c_k, q_k = node(
            current_node.id, LB, current_node.UB, J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, current_node.L0, current_node.L1, M, current_node.is_root,
            min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max
        )

        if result == 0
            println("Node $(current_node.id) pruned")
        elseif result == 1
            println("Optimal integer solution found at node $(current_node.id)")
            return K_star, θ_k, c_k, q_k, LB
        else
            println("Node $(current_node.id) processed, new nodes: ", [n[1] for n in new_nodes])
            println("Updated LB: $LB")
            
            # DFS: Process left child immediately, push right child to stack
            left_l, left_L0, left_L1 = new_nodes[1]  # Left child
            right_l, right_L0, right_L1 = new_nodes[2]  # Right child
            
            # Store left node in node_data
            node_data[left_l] = Node(left_l, UB_l, left_L0, left_L1, false)
            
            # Process left child recursively
            println("Processing left child node $left_l")
            left_result, LB, left_UB, left_new_nodes, K_star, θ_k, c_k, q_k = node(
                left_l, LB, UB_l, J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, left_L0, left_L1, M, false,
                min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max
            )
            
            if left_result == 0
                println("Left child node $left_l pruned")
                # Push right child to stack
                node_data[right_l] = Node(right_l, UB_l, right_L0, right_L1, false)
                push!(nodes, right_l)
                println("Pushed right child node $right_l to stack")
            elseif left_result == 1
                println("Optimal integer solution found at left child node $left_l")
                return K_star, θ_k, c_k, q_k, LB
            else
                println("Left child node $left_l processed, new nodes: ", [n[1] for n in left_new_nodes])
                # Push left child's new nodes to stack
                for (new_l, new_L0, new_L1) in reverse(left_new_nodes)  # Reverse to maintain DFS order
                    node_data[new_l] = Node(new_l, UB_l, new_L0, new_L1, false)
                    push!(nodes, new_l)
                end
                # Push right child to stack
                node_data[right_l] = Node(right_l, UB_l, right_L0, right_L1, false)
                push!(nodes, right_l)
                println("Pushed right child node $right_l to stack")
            end
        end
    end

    println("No feasible solution found")
    return K_star, θ_k, c_k, q_k, LB
end

# Run
K_star, θ_k, c_k, q_k, LB = branch_and_price(
    J, T, K_star, θ_k, c_k, q_k, r_t, u_j, q_j, V_b, γ, Q, e, ρ, ΔSoC, SoC_1, M,
    min_statup, max_statup, min_cpu_time, max_cpu_time, min_periodo_job, max_periodo_job, win_min, win_max
)

# Print final results
println("Final LB: ", LB)
println("Number of profiles per task: ", Dict(j => length(K_star[j]) for j in J))
```
